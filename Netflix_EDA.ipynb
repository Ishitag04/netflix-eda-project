```python
movies = df[df['type']=='Movie']
movies
```

```python
movies['duration_minutes'] = movies['duration'].str.extract('(\d+)').astype(float)
avg_movies = movies['duration_minutes'].mean()
print(f"Average movie duration in minutes = {avg_movies}")
```

```python
TVshow = df[df['type']=='TV Show']
TVshow
```

```python
TVshow = TVshow.fillna(1)
TVshow
```

```python
TVshow['duration_seasons'] = TVshow['duration'].str.extract('(\d+)').astype(float)
avg_seasons = TVshow['duration_seasons'].mean()
print(f"Average TV show duration of seasons = {avg_seasons}")
```
